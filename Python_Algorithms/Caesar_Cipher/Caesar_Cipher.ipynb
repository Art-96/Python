```python
def encrypt(msg, shift):
    
    encrypted_msg = ""
    
    for c in msg:

        # check if character is an uppercase letter
        if c.isupper():

            # find the position in 0-25
            c_unicode = ord(c)

            c_index = ord(c) - ord("A")

            # perform the shift
            new_index = (c_index - shift) % 26

            # convert to new character
            new_unicode = new_index + ord("A")

            new_character = chr(new_unicode)

            # append to encrypted string
            encrypted_msg = encrypted_msg + new_character

        else:
            # since character is not uppercase, leave it as it is
            encrypted_msg += c

    print("Plain text:", msg)

    print("Encrypted text:",encrypted_msg)

#     return encrypted_msg
```

```python
def decrypt(encrypted_msg, shift):

    plain_text = ""

    for c in encrypted_msg:

        # check if character is an uppercase letter
        if c.isupper():

            # find the position in 0-25
            c_unicode = ord(c)

            c_index = ord(c) - ord("A")

            # perform the negative shift
            new_index = (c_index + shift) % 26

            # convert to new character
            new_unicode = new_index + ord("A")

            new_character = chr(new_unicode)

            # append to plain string
            plain_text = plain_text + new_character

        else:

            # since character is not uppercase, leave it as it is
            plain_text += c

    print("Encrypted text:",encrypted_msg)

    print("Decrypted text:",plain_text)
    
#     return clear_msg
```

```python
with open("text.txt", "r") as file:
    Lines = file.readlines()
    n = 3 # defining the shift count
    for line in Lines:
        print(encrypt(line.strip(), n))

print("------------------------------------------------------")
        
with open("text2.txt", "r") as file:
    line = file.readline()
    n = 3 # defining the shift count
    while line:
        print(decrypt(line, n))
        line = file.readline()
```
